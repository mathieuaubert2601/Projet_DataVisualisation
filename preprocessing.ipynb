### III.Préprocessing

```python
df['date_mutation'] = pd.to_datetime(df['date_mutation'], format='%Y-%m-%d')
```

```python
#function to delete column with more than 40% missing data
def delete_too_much_nan(data) : 
    list=[]
    for column in data.columns:
        if(((data[column].isna().sum())*100)/data.shape[0] > 50) :
            list.append(column)
    data = data.drop(columns = list)
    return data

df = delete_too_much_nan(df)
```

```python
#We delete all not necessary columns for our analysis 
df = df.drop(columns=["id_mutation","adresse_numero","adresse_nom_voie","adresse_code_voie","code_nature_culture","nature_culture","code_postal","nom_commune","type_local","numero_disposition"])
```

```python
#We drop all the lines with missing values
df = df.dropna()
```

```python
#We select only the lines with the mutation type equal to "Vente"
df = df[df["nature_mutation"] == "Vente"]
```

```python
#We create labels tranches for the price 
tranches_valeur = [0, 50000, 100000, 200000, 300000, 500000, 1000000, float('inf')]
labels_tranches = [
    "< 50k €",
    "50k - 100k €",
    "100k - 200k €",
    "200k - 300k €",
    "300k - 500k €",
    "500k - 1M €",
    "> 1M €"
]

df['tranche_valeur'] = pd.cut(df['valeur_fonciere'], bins=tranches_valeur, labels=labels_tranches, include_lowest=True)
```

```python
#We create labels tranches for the surface 
surface_bins = [0, 50, 100, 150, 200, 500, np.inf]
surface_labels = ['0-50 m²', '50-100 m²', '100-150 m²', '150-200 m²', '200-500 m²', '>500 m²']
df['plage_surface'] = pd.cut(df['surface_terrain'], bins=surface_bins, labels=surface_labels, include_lowest=True)
```

```python
#Create a column with the price per square metre
df['prix_m2'] = df['valeur_fonciere'] / df['surface_terrain']
```

```python
#Create a colum with the month
df['mois_annee'] = df['date_mutation'].dt.to_period('M')
```

```python
df.to_csv('data2023.csv', index=False)
```
